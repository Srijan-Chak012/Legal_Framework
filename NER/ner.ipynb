```python
# Initialize dictionaries to store entities
entities = {
    'PERSON': {},
    'ORG': {},
    'LAW': {},
    'OTHER': {}
}

# Specify the path to your text file
file_path = '32_ner.txt'

# Open the file and read its contents
with open(file_path, 'r') as file:
    text = file.read()

# Split the text into lines
lines = text.strip().split('\n')

# Iterate over the lines
for line in lines:

    # Check if last word is 'PERSON'
    if line.endswith('PERSON'):
        # Split the line into words
        words = line.split()
        # Get the name
        name = ' '.join(words[:-1])
        
        # If name is not there in dictionary, keep value as 1, otherwise add 1 to previous value
        if name not in entities['PERSON']:
            entities['PERSON'][name] = 1
        else:
            entities['PERSON'][name] += 1


    # Check if last word is 'ORG'
    elif line.endswith('ORG'):
        # Split the line into words
        words = line.split()
        # Get the name
        name = ' '.join(words[:-1])
        
        # If org is not there in dictionary, keep value as 1, otherwise add 1 to previous value
        if name not in entities['ORG']:
            entities['ORG'][name] = 1
        else:
            entities['ORG'][name] += 1

    # Check if last word is 'LAW'
    elif line.endswith('LAW'):
        # Split the line into words
        words = line.split()
        # Get the name
        name = ' '.join(words[:-1])
        
        # If law is not there in dictionary, keep value as 1, otherwise add 1 to previous value
        if name not in entities['LAW']:
            entities['LAW'][name] = 1
        else:
            entities['LAW'][name] += 1

    # Put the rest of the entries into 'OTHER'
    else:
        # Split the line into words
        words = line.split()
        # Get the name
        name = ' '.join(words[:-1])
        
        # If name is not there in dictionary, keep value as 1, otherwise add 1 to previous value
        if name not in entities['OTHER']:
            entities['OTHER'][name] = 1
        else:
            entities['OTHER'][name] += 1


# Write the Persons dictionary into a text file in sorted order
with open('persons_32.txt', 'w') as file:
    # Sort the dictionary according to highest value
    sorted_names = sorted(entities['PERSON'], key=entities['PERSON'].get, reverse=True)
    
    # Iterate over the sorted names
    for name in sorted_names:
        # Make a copy of the name
        temp = name
        # Make everything capital
        name = name.upper()
        # Remove the word 'Shri'
        name = name.replace('SHRI ', '', 1)
        name = name.replace('SRI ', '', 1)
        # Write only letters, remove all punctuation
        name = ''.join([letter for letter in name if letter.isalpha() or letter == ' '])

        # Remove all names <=3 characters
        if len(name) <= 3:
            continue

        # Write the name and its frequency into the file
        file.write(name + ' ' + str(entities['PERSON'][temp]) + '\n')

# Write the Organizations dictionary into a text file
with open('orgs_32.txt', 'w') as file:

    # Sort the dictionary according to highest value
    sorted_names = sorted(entities['ORG'], key=entities['ORG'].get, reverse=True)

    # Print sorted names and values
    for name in sorted_names:
        # Make a copy of the name
        temp = name
        # Make everything capital
        name = name.upper()
        # Remove the word 'Shri'
        name = name.replace('SHRI ', '', 1)
        name = name.replace('SRI ', '', 1)
        # Write only letters, remove all punctuation
        name = ''.join([letter for letter in name if letter.isalpha() or letter == ' '])

        # Remove all names <=3 characters
        if len(name) <= 3:
            continue

        # Write the name and its frequency into the file
        file.write(name + ' ' + str(entities['ORG'][temp]) + '\n')
    
# Write the Laws dictionary into a text file
with open('laws_32.txt', 'w') as file:
    # Sort the dictionary according to highest value
    sorted_names = sorted(entities['LAW'], key=entities['LAW'].get, reverse=True)

    # Print sorted names and values
    for name in sorted_names:
        # Make a copy of the name
        temp = name
        # Make everything capital
        name = name.upper()
        # Remove the word 'Shri'
        name = name.replace('SHRI ', '', 1)
        name = name.replace('SRI ', '', 1)
        # Write only letters, remove all punctuation
        name = ''.join([letter for letter in name if letter.isalpha() or letter == ' '])

        # Remove all names <=3 characters
        if len(name) <= 3:
            continue

        # Write the name and its frequency into the file
        file.write(name + ' ' + str(entities['LAW'][temp]) + '\n')
```
